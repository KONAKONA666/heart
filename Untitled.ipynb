```python
import pandas as pd
import numpy as np
import scipy.stats as ss


%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def cramers_corrected_stat(confusion_matrix):
    """ calculate Cramers V statistic for categorial-categorial association.
        uses correction from Bergsma and Wicher, 
        Journal of the Korean Statistical Society 42 (2013): 323-328
    """
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = phi2 - ((k-1)*(r-1))/(n-1)    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr /(kcorr-1))
```

```python
data = pd.read_csv('Heart_Disease_Data.csv')
data.info()
```

```python
data['thal'] = data[data['thal'].notna()].dropna()  
data['ca'] = data[data['ca'].notna()].dropna()  
```

```python
data['thal'] = data['thal'].astype('int64')
data['ca']   = data['ca'].astype('int64')
data = data.astype('float64')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier


X_train, X_test, y_train, y_test = train_test_split(data.drop(['pred_attribute'], axis=1), data['pred_attribute'], test_size=0.35, random_state=50)
clf_ = RandomForestClassifier(n_estimators=24, max_depth=15, random_state=10, n_jobs=-1).fit(X_train, y_train)
import pickle

pickle.dump(clf_, open("model1.sav", 'wb'))
```
